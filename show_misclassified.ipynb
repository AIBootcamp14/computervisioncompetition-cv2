```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import math

# 오분류 이미지가 있는 폴더 경로
folder_path = './misclassified_images'

# 이미지를 저장할 출력 폴더
output_folder = './misclassified_montage'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 이미지 파일 목록 가져오기
image_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

# 이미지들을 25개씩 묶어서 저장
batch_size = 25
num_batches = math.ceil(len(image_files) / batch_size)

for i in range(num_batches):
    start_index = i * batch_size
    end_index = min((i + 1) * batch_size, len(image_files))
    batch_files = image_files[start_index:end_index]
    
    fig = plt.figure(figsize=(20, 20))
    
    for j, image_path in enumerate(batch_files):
        ax = fig.add_subplot(5, 5, j + 1)
        img = mpimg.imread(image_path)
        ax.imshow(img)
        ax.set_title(os.path.basename(image_path), fontsize=8)
        ax.axis('off')
        
    plt.tight_layout()
    output_filename = f'misclassified_batch_{i+1}.png'
    plt.savefig(os.path.join(output_folder, output_filename))
    plt.close(fig) # 메모리 해제

print(f"총 {num_batches}개의 오분류 이미지 묶음 파일을 '{output_folder}'에 저장했습니다.")
```
